```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from statsmodels.stats.stattools import durbin_watson


df = pd.read_csv('APMM8-assignment2-dataset15.csv')


X = df[['X']].values  # Reshape to 2D array for sklearn
y = df['y'].values


model = LinearRegression()
model.fit(X, y)


intercept = model.intercept_
coefficient = model.coef_[0]

# Calculate predictions and R-squared
y_pred = model.predict(X)

print(f"Intercept: {intercept:.3f}")
print(f"Coefficient: {coefficient:.3f}")

```

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['X'], df['y'], alpha=0.7)
plt.xlabel('X')
plt.ylabel('y')
plt.title('X vs Y Scatter Plot')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Extract X and y values
X = df[['X']].values
y = df['y'].values

# Perform simple linear regression
model = LinearRegression()
model.fit(X, y)

# Calculate residuals
y_pred = model.predict(X)
residuals = y - y_pred

# Calculate Durbin-Watson statistic
dw_statistic = durbin_watson(residuals)

print(f"Durbin-Watson Statistic: {dw_statistic:.3f}")
```

```python
# Create histogram of residuals
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, alpha=0.7, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.grid(True, alpha=0.3)
plt.show()

```
