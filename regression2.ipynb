```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
df = pd.read_csv('APMM8-assignment2-dataset10.csv')

# Extract X and y values
X = df[['X']].values  # Reshape to 2D array for sklearn
y = df['y'].values

# Perform simple linear regression
model = LinearRegression()
model.fit(X, y)

# Get the intercept and coefficient
intercept = model.intercept_
coefficient = model.coef_[0]

# Calculate predictions and R-squared
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print(f"Intercept: {intercept:.3f}")
print(f"Coefficient: {coefficient:.3f}")
print(f"R-squared: {r2:.3f}")

# Alternative method using numpy for verification
X_with_intercept = np.column_stack([np.ones(len(X)), X])
coefficients_np = np.linalg.lstsq(X_with_intercept, y, rcond=None)[0]
print(f"\nVerification with numpy:")
print(f"Intercept: {coefficients_np[0]:.3f}")
print(f"Coefficient: {coefficients_np[1]:.3f}")
X_with_intercept = np.column_stack([np.ones(len(X)), X.ravel()])
coefficients_np = np.linalg.lstsq(X_with_intercept, y, rcond=None)[0]
print(f"Verification with numpy - Weight for X: {coefficients_np[1]:.3f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy import stats

# Load the dataset
df = pd.read_csv('APMM8-assignment2-dataset10.csv')

# Extract X and y values
X = df[['X']].values
y = df['y'].values

# Perform simple linear regression
model = LinearRegression()
model.fit(X, y)

# Calculate residuals
y_pred = model.predict(X)
residuals = y - y_pred

# Create histogram of residuals
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, alpha=0.7, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.grid(True, alpha=0.3)
plt.show()

# Perform Shapiro-Wilk test for normality
shapiro_stat, shapiro_p = stats.shapiro(residuals)
print(f"Shapiro-Wilk test p-value: {shapiro_p:.6f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('APMM8-assignment2-dataset11.csv')

# Extract X and y
X = data['X'].values
y = data['y'].values

# Perform linear regression
n = len(X)
sum_x = np.sum(X)
sum_y = np.sum(y)
sum_xy = np.sum(X * y)
sum_x2 = np.sum(X**2)

# Calculate slope and intercept
b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
b0 = (sum_y - b1 * sum_x) / n

# Calculate fitted values and residuals
fitted_values = b0 + b1 * X
residuals = y - fitted_values

# Create residuals vs fitted values plot
plt.figure(figsize=(10, 6))
plt.scatter(fitted_values, residuals, alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values Plot')
plt.grid(True, alpha=0.3)
plt.show()

# Print some statistics
print(f"Mean of residuals: {np.mean(residuals):.6f}")
print(f"Standard deviation of residuals: {np.std(residuals):.3f}")
```
