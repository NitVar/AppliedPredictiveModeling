```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

def analyze_dataset(filename):
    """Analyze a dataset and check linearity assumption"""
    
    # Load the dataset
    df = pd.read_csv(filename)
    
    print(f"\n{'='*60}")
    print(f"ANALYZING: {filename}")
    print(f"{'='*60}")
    
    # Prepare the data for regression
    X = df[['X1', 'X2', 'X3', 'X4']]  # Features
    y = df['y']  # Target variable
    
    # Perform multiple linear regression
    model = LinearRegression()
    model.fit(X, y)
    
    # Make predictions
    y_pred = model.predict(X)
    
    # Calculate residuals
    residuals = y - y_pred
    
    # Calculate metrics
    r2 = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    
    print(f"R-squared: {r2:.4f}")
    print(f"RMSE: {rmse:.4f}")
    
    # Create residuals vs fitted values plot
    plt.figure(figsize=(10, 6))
    plt.scatter(y_pred, residuals, alpha=0.6)
    plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    plt.title(f'Residuals vs Fitted Values - {filename}')
    plt.grid(True, alpha=0.3)
    
    # Add statistics to the plot
    plt.text(0.05, 0.95, f'R = {r2:.4f}', transform=plt.gca().transAxes, 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
    plt.text(0.05, 0.88, f'RMSE = {rmse:.4f}', transform=plt.gca().transAxes,
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    # Check linearity assumption
    correlation = np.corrcoef(y_pred, residuals)[0, 1]
    print(f"Correlation between fitted values and residuals: {correlation:.4f}")
    
    if abs(correlation) < 0.1:
        print(" Linearity assumption appears to be SATISFIED")
        print("  - Residuals are randomly scattered around zero")
        print("  - No clear patterns in the residuals vs fitted plot")
    else:
        print(" Linearity assumption may be VIOLATED")
        print("  - Residuals show patterns or systematic behavior")
        print("  - Relationship may not be purely linear")
    
    return r2, rmse, correlation

# Analyze all datasets
datasets = [
    'APMM8-assignment2-dataset21.csv',
    'APMM8-assignment2-dataset22.csv', 
    'APMM8-assignment2-dataset23.csv'
]

results = {}
for dataset in datasets:
    try:
        r2, rmse, corr = analyze_dataset(dataset)
        results[dataset] = {'R2': r2, 'RMSE': rmse, 'Correlation': corr}
    except FileNotFoundError:
        print(f"File {dataset} not found. Skipping...")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Set style for better looking plots
plt.style.use('default')

# Load the dataset
df = pd.read_csv('APMM8-assignment2-dataset21.csv')

# Prepare the data for regression
X = df[['X1', 'X2', 'X3', 'X4']]  # Features
y = df['y']  # Target variable

# Perform multiple linear regression
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Calculate residuals
residuals = y - y_pred

# Calculate metrics for residuals
residual_mean = np.mean(residuals)
residual_std = np.std(residuals)
skewness = stats.skew(residuals)
kurtosis = stats.kurtosis(residuals)

# Perform normality test
shapiro_stat, shapiro_p = stats.shapiro(residuals)

# Create histogram of residuals
plt.figure(figsize=(12, 8))

# Plot histogram
n, bins, patches = plt.hist(residuals, bins=30, alpha=0.7, color='skyblue', 
                           edgecolor='black', linewidth=0.5, density=True, 
                           label='Residuals')

# Add normal distribution curve for comparison
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, residual_mean, residual_std)
plt.plot(x, p, 'red', linewidth=2, label='Normal distribution')

# Add vertical line at mean
plt.axvline(residual_mean, color='green', linestyle='--', linewidth=2, 
           label=f'Mean: {residual_mean:.4f}')

# Add labels and title
plt.xlabel('Residuals', fontsize=14, fontweight='bold')
plt.ylabel('Density', fontsize=14, fontweight='bold')
plt.title('Histogram of Residuals with Normal Distribution Curve\nDataset: APMM8-assignment2-dataset21.csv', 
          fontsize=16, fontweight='bold', pad=20)

# Add grid
plt.grid(True, alpha=0.3)

# Add statistics box
stats_text = f'Mean: {residual_mean:.6f}\nStd Dev: {residual_std:.4f}\nSkewness: {skewness:.4f}\nKurtosis: {kurtosis:.4f}\nShapiro-Wilk p-value: {shapiro_p:.6f}\nN = {len(residuals)}'
plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes, 
         bbox=dict(boxstyle="round,pad=0.5", facecolor="white", alpha=0.9),
         verticalalignment='top', fontsize=11, fontfamily='monospace')

# Add legend
plt.legend(loc='upper right', framealpha=0.9)

# Adjust layout and save
plt.tight_layout()

# Save the plot
plt.savefig('residuals_histogram_dataset21.png', 
            dpi=300, 
            bbox_inches='tight', 
            facecolor='white',
            edgecolor='none')

plt.show()

# Print detailed analysis
print("="*70)
print("NORMALITY ANALYSIS OF RESIDUALS - Dataset 21")
print("="*70)
print(f"Residual Statistics:")
print(f"  Mean: {residual_mean:.6f}")
print(f"  Standard Deviation: {residual_std:.4f}")
print(f"  Skewness: {skewness:.4f}")
print(f"  Kurtosis: {kurtosis:.4f}")
print(f"  Shapiro-Wilk test statistic: {shapiro_stat:.4f}")
print(f"  Shapiro-Wilk p-value: {shapiro_p:.6f}")

print("\nNormality Assessment:")
print("-" * 50)

# Assess normality based on statistics
normality_indicators = []

# Check skewness (should be close to 0 for normality)
if abs(skewness) < 0.5:
    normality_indicators.append(" Skewness is close to zero")
else:
    normality_indicators.append(f" Skewness ({skewness:.4f}) suggests asymmetry")

# Check kurtosis (should be close to 0 for normality)
if abs(kurtosis) < 1:
    normality_indicators.append(" Kurtosis is close to normal")
else:
    normality_indicators.append(f" Kurtosis ({kurtosis:.4f}) suggests non-normal tails")

# Check Shapiro-Wilk test (p > 0.05 suggests normality)
if shapiro_p > 0.05:
    normality_indicators.append(" Shapiro-Wilk test fails to reject normality")
else:
    normality_indicators.append(" Shapiro-Wilk test rejects normality")

# Print all indicators
for indicator in normality_indicators:
    print(indicator)

# Overall conclusion
print(f"\nOVERALL CONCLUSION:")
if shapiro_p > 0.05 and abs(skewness) < 0.5 and abs(kurtosis) < 1:
    print(" The residuals appear to be approximately normally distributed")
else:
    print(" The residuals do not appear to be normally distributed")

# Additional Q-Q plot for visual confirmation
plt.figure(figsize=(10, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals\n(Additional Normality Check)', 
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('qq_plot_residuals_dataset21.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Set style for better looking plots
plt.style.use('default')
sns.set_palette("viridis")

# Load the dataset
df = pd.read_csv('APMM8-assignment2-dataset21.csv')

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create the correlation heatmap
plt.figure(figsize=(10, 8))

# Create heatmap
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))  # Mask upper triangle
heatmap = sns.heatmap(correlation_matrix, 
                     mask=mask,
                     annot=True, 
                     fmt=".3f", 
                     cmap='RdBu_r',
                     center=0,
                     square=True,
                     cbar_kws={'shrink': 0.8})

plt.title('Correlation Matrix Heatmap\nDataset: APMM8-assignment2-dataset21.csv', 
          fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()

# Save the plot
plt.savefig('correlation_matrix_dataset21.png', 
            dpi=300, 
            bbox_inches='tight', 
            facecolor='white',
            edgecolor='none')

plt.show()

# Calculate VIF (Variance Inflation Factor) for multicollinearity assessment
X = df[['X1', 'X2', 'X3', 'X4']]
X_with_const = sm.add_constant(X)  # Add constant for VIF calculation

vif_data = pd.DataFrame()
vif_data["Variable"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) 
                   for i in range(X_with_const.shape[1])]

# Print detailed analysis
print("="*70)
print("MULTICOLLINEARITY ANALYSIS - Dataset 21")
print("="*70)

print("\nCORRELATION MATRIX:")
print("-" * 50)
print(correlation_matrix.round(4))

print("\nVARIANCE INFLATION FACTORS (VIF):")
print("-" * 50)
print(vif_data.round(4))

print("\nMULTICOLLINEARITY ASSESSMENT:")
print("-" * 50)

# Assess multicollinearity
high_corr_count = 0
high_vif_count = 0

# Check for high correlations (absolute value > 0.7)
print("\nHigh Correlations (|r| > 0.7):")
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = abs(correlation_matrix.iloc[i, j])
        if corr_val > 0.7:
            print(f" {correlation_matrix.columns[i]} vs {correlation_matrix.columns[j]}: {correlation_matrix.iloc[i, j]:.4f}")
            high_corr_count += 1

if high_corr_count == 0:
    print(" No high correlations found between predictor variables")

# Check for high VIF values (VIF > 10 indicates severe multicollinearity)
print("\nVIF Assessment:")
for _, row in vif_data.iterrows():
    if row['Variable'] != 'const':  # Skip constant
        if row['VIF'] > 10:
            print(f" {row['Variable']}: VIF = {row['VIF']:.2f} (Severe multicollinearity)")
            high_vif_count += 1
        elif row['VIF'] > 5:
            print(f" {row['Variable']}: VIF = {row['VIF']:.2f} (Moderate multicollinearity)")
            high_vif_count += 1
        else:
            print(f" {row['Variable']}: VIF = {row['VIF']:.2f} (Acceptable)")

# Overall conclusion
print(f"\nOVERALL CONCLUSION:")
if high_corr_count == 0 and high_vif_count == 0:
    print(" The dataset does NOT exhibit multicollinearity")
    print("  - All inter-predictor correlations are low")
    print("  - All VIF values are within acceptable range")
else:
    print(" The dataset exhibits multicollinearity")
    if high_corr_count > 0:
        print(f"  - Found {high_corr_count} high correlation(s) between predictors")
    if high_vif_count > 0:
        print(f"  - Found {high_vif_count} variable(s) with elevated VIF values")

# Additional: Condition Number for multicollinearity
X_matrix = X.values
condition_number = np.linalg.cond(X_matrix)
print(f"\nCondition Number: {condition_number:.2f}")

if condition_number > 30:
    print(" High condition number suggests potential multicollinearity")
else:
    print(" Condition number suggests no significant multicollinearity")

# Create a scatter plot matrix to visualize relationships
print("\nGenerating scatter plot matrix...")
scatter_matrix = pd.plotting.scatter_matrix(df, figsize=(12, 10), alpha=0.6, diagonal='hist')
plt.suptitle('Scatter Plot Matrix - All Variables', fontsize=16, fontweight='bold', y=0.95)
plt.tight_layout()
plt.savefig('scatter_matrix_dataset21.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import scipy.stats as stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

def analyze_dataset(filename):
    """
    Comprehensive regression analysis for a dataset
    Performs: regression, residual analysis, multicollinearity check
    """
    
    # Set style for plots
    plt.style.use('default')
    sns.set_palette("viridis")
    
    # Load the dataset
    df = pd.read_csv(filename)
    
    print(f"\n{'='*80}")
    print(f"COMPREHENSIVE ANALYSIS: {filename}")
    print(f"{'='*80}")
    
    # Prepare data for regression
    X = df[['X1', 'X2', 'X3', 'X4']]
    y = df['y']
    
    # 1. PERFORM MULTIPLE LINEAR REGRESSION
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    residuals = y - y_pred
    
    # Calculate regression metrics
    r2 = r2_score(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    
    print(f"\nREGRESSION RESULTS:")
    print(f"R-squared: {r2:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"Coefficients: {model.coef_}")
    print(f"Intercept: {model.intercept_:.4f}")
    
    # 2. RESIDUALS VS FITTED PLOT
    plt.figure(figsize=(12, 10))
    
    plt.subplot(2, 2, 1)
    plt.scatter(y_pred, residuals, alpha=0.6, color='blue')
    plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    plt.title('Residuals vs Fitted Values')
    plt.grid(True, alpha=0.3)
    
    # Check linearity assumption
    corr_fitted_residuals = np.corrcoef(y_pred, residuals)[0, 1]
    plt.text(0.05, 0.95, f'R = {r2:.4f}\nCorr = {corr_fitted_residuals:.4f}', 
             transform=plt.gca().transAxes, 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.9))
    
    # 3. HISTOGRAM OF RESIDUALS
    plt.subplot(2, 2, 2)
    residual_mean = np.mean(residuals)
    residual_std = np.std(residuals)
    skewness = stats.skew(residuals)
    kurtosis = stats.kurtosis(residuals)
    shapiro_stat, shapiro_p = stats.shapiro(residuals)
    
    # Plot histogram
    n, bins, patches = plt.hist(residuals, bins=30, alpha=0.7, color='skyblue', 
                               edgecolor='black', density=True)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, residual_mean, residual_std)
    plt.plot(x, p, 'red', linewidth=2)
    plt.axvline(residual_mean, color='green', linestyle='--', linewidth=2)
    
    plt.xlabel('Residuals')
    plt.ylabel('Density')
    plt.title('Histogram of Residuals')
    plt.grid(True, alpha=0.3)
    
    stats_text = f'Mean: {residual_mean:.4f}\nStd: {residual_std:.4f}\nSkew: {skewness:.4f}\nKurt: {kurtosis:.4f}\nShapiro p: {shapiro_p:.4f}'
    plt.text(0.05, 0.95, stats_text, transform=plt.gca().transAxes,
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.9),
             verticalalignment='top', fontsize=9)
    
    # 4. CORRELATION MATRIX HEATMAP
    plt.subplot(2, 2, 3)
    correlation_matrix = df.corr()
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt=".3f", 
                cmap='RdBu_r', center=0, square=True, cbar_kws={'shrink': 0.8})
    plt.title('Correlation Matrix')
    
    # 5. Q-Q PLOT
    plt.subplot(2, 2, 4)
    stats.probplot(residuals, dist="norm", plot=plt)
    plt.title('Q-Q Plot of Residuals')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'comprehensive_analysis_{filename.replace(".csv", "")}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()
    
    # 6. MULTICOLLINEARITY ANALYSIS
    print(f"\nMULTICOLLINEARITY ANALYSIS:")
    print("-" * 50)
    
    # Calculate VIF
    X_with_const = sm.add_constant(X)
    vif_data = pd.DataFrame()
    vif_data["Variable"] = X_with_const.columns
    vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) 
                       for i in range(X_with_const.shape[1])]
    
    print("Variance Inflation Factors (VIF):")
    print(vif_data.round(4))
    
    # Check for high correlations between predictors
    high_corr_count = 0
    predictor_corrs = correlation_matrix[['X1', 'X2', 'X3', 'X4']].loc[['X1', 'X2', 'X3', 'X4']]
    
    print("\nInter-predictor correlations:")
    for i in range(len(predictor_corrs.columns)):
        for j in range(i+1, len(predictor_corrs.columns)):
            corr_val = abs(predictor_corrs.iloc[i, j])
            if corr_val > 0.7:
                high_corr_count += 1
                print(f" {predictor_corrs.columns[i]} vs {predictor_corrs.columns[j]}: {predictor_corrs.iloc[i, j]:.4f}")
    
    # 7. ASSUMPTIONS ASSESSMENT
    print(f"\nASSUMPTIONS ASSESSMENT:")
    print("-" * 50)
    
    # Linearity assessment
    if abs(corr_fitted_residuals) < 0.1:
        print(" Linearity: SATISFIED (residuals randomly scattered)")
    else:
        print(" Linearity: VIOLATED (pattern in residuals)")
    
    # Normality assessment
    if shapiro_p > 0.05:
        print(" Normality: SATISFIED (residuals normally distributed)")
    else:
        print(" Normality: VIOLATED (residuals not normal)")
    
    # Multicollinearity assessment
    max_vif = vif_data[vif_data['Variable'] != 'const']['VIF'].max()
    if max_vif < 5 and high_corr_count == 0:
        print(" Multicollinearity: NO ISSUES (low correlations & VIF)")
    elif max_vif < 10:
        print(" Multicollinearity: MODERATE (elevated VIF)")
    else:
        print(" Multicollinearity: SEVERE (high VIF)")
    
    # Return results for comparison
    results = {
        'filename': filename,
        'r2': r2,
        'rmse': rmse,
        'shapiro_p': shapiro_p,
        'max_vif': max_vif,
        'max_predictor_corr': predictor_corrs.values[np.triu_indices_from(predictor_corrs.values, k=1)].max(),
        'linearity_ok': abs(corr_fitted_residuals) < 0.1,
        'normality_ok': shapiro_p > 0.05,
        'multicollinearity_ok': max_vif < 5 and high_corr_count == 0
    }
    
    return results

# USAGE EXAMPLE:
if __name__ == "__main__":
    # Analyze multiple datasets
    datasets = [
        'APMM8-assignment2-dataset23.csv'
        
        # Add more datasets as needed
    ]
    
    all_results = []
    
    for dataset in datasets:
        try:
            results = analyze_dataset(dataset)
            all_results.append(results)
        except FileNotFoundError:
            print(f"File {dataset} not found. Skipping...")
        except Exception as e:
            print(f"Error analyzing {dataset}: {e}")
    
    # Print summary comparison
    if all_results:
        print(f"\n{'='*80}")
        print("SUMMARY COMPARISON ACROSS DATASETS")
        print(f"{'='*80}")
        
        summary_df = pd.DataFrame(all_results)
        print(summary_df.round(4))
```
